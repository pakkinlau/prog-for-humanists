# Section 1 - Sign up a GitHub account and Getting GitHub Student Developer Pack <a id='section1'></a>

GitHub is a popular platform for version control and collaborative software development. As a student, you can take advantage of GitHub's Student Developer Pack, which provides access to a variety of tools and resources for free. In this Jupyter Notebook guide, we will walk you through the process of registering on GitHub and obtaining the GitHub Student Developer Pack.

#### Prerequisites:

1. **A valid student email address** (ending with `.edu`) that you can verify.
2. **Basic understanding of Git and version control concepts.**

### Steps:

#### 1. **Create a GitHub Account:**

If you already have a GitHub account, you can skip this step. Otherwise, go to [GitHub](https://github.com/) and sign up for a new account using your student email address.

#### 2. **Verify Your Student Status:**

After creating your GitHub account, go to the [GitHub Student Developer Pack](https://education.github.com/pack) page. Click on the "Get Your Pack" button and follow the instructions to verify your student status. You will need to provide your school-issued email address for verification.

Alt text

#### 3. **Wait for Verification:**

Once you have submitted your verification request, GitHub will review your application. This process may take some time, so be patient.

![400](image-3.png)

#### 4. **Explore the GitHub Student Developer Pack:**

Once your student status is verified, you will receive an email confirmation. You can then access the GitHub Student Developer Pack/

Alt text

Here, you can find various offers and tools provided by GitHub partners. These can include free access to hosting services, domain names, development tools, and more.

#### 5. **Start Using the Tools:**

Once you have redeemed the offers, you can start using the tools and resources provided by the GitHub Student Developer Pack to enhance your coding projects and collaborate with others more effectively.
